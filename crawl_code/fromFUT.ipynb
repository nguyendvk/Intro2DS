```python
import requests
import requests_cache
import json
import pandas as pd
import csv
```

```python
sleep_time = 1
requests_cache.install_cache(filter_fn=lambda response: response.text.find('"incomplete_results":true') == -1)
```

```python
def get_request(url):
    r = requests.get(url)
    # print(r.ok)
    while not r.ok or r.text.find('"incomplete_results":true') > -1:
        if not r.ok:
            time.sleep(sleep_time)
        r = requests.get(url)
    return r
```

```python
url = 'https://www.easports.com/fifa/ultimate-team/api/fut/item?page='
r = json.loads(get_request(url+'1').text)
num_pages = r['totalPages']
items = []
for i in range(num_pages):
    r = json.loads(get_request(url+str(i)).text)
#     print(len(r['items']))
    items += r['items']
```

```python
columns = items[0].keys()
with open('../data/FUT.csv', 'w', encoding='UTF8' ,newline='')  as output_file:
    dict_writer = csv.DictWriter(output_file, columns)
    dict_writer.writeheader()
    dict_writer.writerows(items)
```
