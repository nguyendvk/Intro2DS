This notebook get all leagues of 5 countries England, France, Spain, Italy, Germany and save them to 'leagues.csv' file

```python
import requests
import requests_cache
import json
import pandas as pd
```

```python
requests_cache.install_cache(filter_fn=lambda response: response.text.find('"incomplete_results":true') == -1)
```

```python
countries = ['England', 'France', 'Spain', 'Italy', 'Germany']
all_leagues = []
for country in countries:
    url = 'https://www.thesportsdb.com/api/v1/json/1/search_all_leagues.php?c=' + country
    leagues = json.loads(requests.get(url).text)['countrys']
    for l in leagues:
        l['country'] = country
    all_leagues += leagues
len(all_leagues)
```

```python
# save all_leagues to 'leagues.csv'
import csv
keys = all_leagues[0].keys()
with open('data/leagues.csv', 'w', encoding='UTF8' ,newline='')  as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(all_leagues)
```
